```python
'''
This is a simple calculator created using Python. The program asks the user to
input two numbers and outputs the sum, difference, product, and quotient of the
two numbers.
'''

# Prompt the user for two integers and store them as variables
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Store the operations as variables
sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

# Print the required outputs
print(num1, " + ", num2, " = ", sum)
print(num1, " - ", num2, " = ", difference)
print(num1, " * ", num2, " = ", product)
print(num1, " / ", num2, " = ", quotient)
```
